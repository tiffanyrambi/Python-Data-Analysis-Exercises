Merging DataFrames:
merge, join, concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\brend\Dropbox\My PC (LAPTOP-9LJ9T83N)\Downloads\PandasYouTubeSeries-main\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\brend\Dropbox\My PC (LAPTOP-9LJ9T83N)\Downloads\PandasYouTubeSeries-main\LOTR 2.csv")
df2
```

```python
# INNER JOIN USING MERGE
df1.merge(df2)
```

```python
# INNER JOIN USING MERGE - SPECIFYING THE JOIN
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])

# different ways
# df1.merge(df2, how = 'inner', on = 'FellowshipID')
# df1.merge(df2, how = 'inner', on = 'FellowshipID', suffixes = ('_1','_2'))
```

```python
# OUTER JOIN
df1.merge(df2, how = 'outer')
```

```python
# LEFT JOIN
df1.merge(df2, how = 'left')
```

```python
# RIGHT JOIN
df1.merge(df2, how = 'right')
```

```python
# CROSS JOIN
df1.merge(df2, how = 'cross')
```

```python
# JOIN USING JOIN FUNCTION
```

```python
df1.join(df2, on = 'FellowshipID', how= 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4

# specify the join
#df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'inner')
```

```python
# CONCATENATE
pd.concat([df1,df2])
```

```python
# CONCATE - INNER JOIN
# only shows the same columns on both data
pd.concat([df1,df2], join = 'inner')
```

```python
# CONCATE - OUTER JOIN
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

```python
# APPEND
df1.append(df2)

# depracted!!!!
```
