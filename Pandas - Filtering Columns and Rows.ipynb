Filtering and Ordering

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\brend\Dropbox\My PC (LAPTOP-9LJ9T83N)\Downloads\PandasYouTubeSeries-main\world_population.csv")
df
```

```python
# Filter Rank <= 10
df[df['Rank'] <= 10]
```

```python
# isin() function
specific_countries = ['Bangladesh', 'Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
# contains() function
df[df['Country'].str.contains('Republic')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
# filter columns
# df2.filter(items = ['Continent', 'CCA3'])

# changing axis
# Default axis = 1
# Axis 0 = X axis
# Axis 1 = Y axis
df2.filter(items = ['Continent', 'CCA3'], axis = 0)
```

```python
df2.filter(items = ['Indonesia'], axis = 0)
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
df2.loc['United States']
# case sensitive
```

```python
df2.iloc[3]
# case sensitive
```

```python
# Sort Values
df[df['Rank'] <=10].sort_values(by='Rank')

# Ascending/Descending
# df[df['Rank'] <=10].sort_values(by='Rank', ascending=False)
```

```python
# Sorting based on multiple columns
df[df['Rank'] <=10].sort_values(by=['Continent', 'Country'], ascending=True)

# Sorting: multiple columns & multiple asc/desc
df[df['Rank'] <=10].sort_values(by=['Continent', 'Country'], ascending=[False,True])
```
